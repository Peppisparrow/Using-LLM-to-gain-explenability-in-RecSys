```python
import pandas as pd
ratings = pd.read_csv('ratings.csv')
ratings
```

```python
ratings.rename(columns={'book_id': 'app_id'}, inplace=True)
```

```python
import numpy as np
users = ratings['user_id'].unique().tolist()
sampled_users = np.random.choice(users, size=1000, replace=False)
sampled_ratings = ratings[ratings['user_id'].isin(sampled_users)]
sampled_ratings
```

```python
#mostra qualche metrica del tipo numero massimo di interazioni per utente, numero minimo, media, mediana
ratings.groupby('user_id').size().describe(),ratings.groupby('app_id').size().describe()
```

```python
sampled_ratings.groupby('user_id').size().describe(),sampled_ratings.groupby('app_id').size().describe()
```

```python
#calcola e confronta la sparsita del dataset originale e del sottoinsieme
len(ratings) / (ratings['user_id'].nunique() * ratings['app_id'].nunique()), len(sampled_ratings) / (sampled_ratings['user_id'].nunique() * sampled_ratings['app_id'].nunique())
```

```python
sampled_ratings.to_csv('recommendations_1000users.csv', index=False)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
train_recommendations, test_recommendations = train_test_split(sampled_ratings, test_size=0.1, random_state=42)
```
